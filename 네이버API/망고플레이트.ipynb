```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

keyword = "서울"
url = f"https://www.mangoplate.com/search/{keyword}"
headers = {'User-Agent' : 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
print('응답: ', response)
```

```python
print(response.text)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
data = soup.select("li.server_render_search_result_item > div.list-restaurant-item")
print(data[0])
```

```python
for item in data:
    image= item.select_one('img').get('data-original')
    link= item.select_one('a').get('href')
    title= item.select_one('h2.title').text.replace('\n', '')
    rating= item.select_one('strong.search_point').text
    category= item.select_one('p.etc').text
    view= item.select_one('span.view_count').text
    review= item.select_one('span.review_count').text

    print("썸네일 주소: ", image)
    print("url       : ", link)
    print("가게 이름  : ", title)
    print("평점      : ", rating)
    print("가게 종류  : ", category)
    print("조회수     : ", view)
    print("리뷰 수    : ", review)
    print()
```

```python
thumbnails = []
urls = []
titles = []
ratings = []
categories = []
views = []
reviews = []

for item in data:
    image= item.select_one('img').get('data-original')
    link= item.select_one('a').get('href')
    title= item.select_one('h2.title').text.replace('\n', '')
    rating= item.select_one('strong.search_point').text
    category= item.select_one('p.etc').text
    view= item.select_one('span.view_count').text
    review= item.select_one('span.review_count').text

    thumbnails.append(image)
    urls.append(link)
    titles.append(title)
    ratings.append(rating)
    categories.append(category)
    views.append(view)
    reviews.append(review)

data_dict = {
    "썸네일 주소": thumbnails,
    "url": urls,
    "가게 이름": titles,
    "평점": ratings,
    "가게 종류": categories,
    "조회수": views,
    "리뷰 수": reviews  
}

# 데이터프레임 생성
df = pd.DataFrame(data_dict)

# 데이터프레임 출력
print(df.info())
```

```python
df.to_csv("망고플레이트_서울맛집20.csv", index=False)
```
