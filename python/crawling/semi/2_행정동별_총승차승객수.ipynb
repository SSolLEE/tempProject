```python
import requests
import json
import pandas as pd

SERVICE_KEY = '69576567536c736f3133357542784147'

data = None
for j in range(1, 200):
    url = f'http://openapi.seoul.go.kr:8088/{SERVICE_KEY}/Json/tpssPassengerCnt/{1+(j-1)*1000}/{j*1000}/'
    # print(url)
    req = requests.get(url)
    content = req.json()
    con = content['tpssPassengerCnt']['row']
    result = pd.DataFrame(con)
    data = pd.concat([data, result])
data = data.reset_index(drop=True)
data.info()
    
```

```python
import requests
import json
import pandas as pd

SERVICE_KEY = '69576567536c736f3133357542784147'

df = None
for j in range(1, 3):
    url = f'http://openapi.seoul.go.kr:8088/{SERVICE_KEY}/Json/districtEmd/{1+(j-1)*1000}/{j*1000}/'
    # print(url)
    req = requests.get(url)
    content = req.json()
    con = content['districtEmd']['row']
    result = pd.DataFrame(con)
    df = pd.concat([df, result])
df = df.reset_index(drop=True)
df.info()
    
```
