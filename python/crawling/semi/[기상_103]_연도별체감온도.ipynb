# 연도별 체감온도 
- 여름철(5-9월), 겨울철(10-익년4월) 이렇게 데이터 제공
- 관측값 기반
  
- 겨울철 데이터 일부 결측값 존재

```python
import pandas as pd

temp_2021_1 = pd.read_csv("./raw/STCS_체감온도_2021_01_04.csv")
temp_2021_2 = pd.read_csv("./raw/STCS_체감온도_2021_05_09.csv")
temp_2021_3 = pd.read_csv("./raw/STCS_체감온도_2021_10_12.csv")
temp_2022_1 = pd.read_csv("./raw/STCS_체감온도_2022_01_04.csv")
temp_2022_2 = pd.read_csv("./raw/STCS_체감온도_2022_05_09.csv")
temp_2022_3 = pd.read_csv("./raw/STCS_체감온도_2022_10_12.csv")
temp_2023_1 = pd.read_csv("./raw/STCS_체감온도_2023_01_04.csv")
temp_2023_2 = pd.read_csv("./raw/STCS_체감온도_2023_05_09.csv")
```

## 1년짜리 데이터 만들기

### 컬럼 정리
- index 0 삭제
- index 1 컬럼명 지정 후 삭제

```python
temp_2021_1.drop(temp_2021_1.index[0], inplace=True)
temp_2021_1 = temp_2021_1.rename(columns=temp_2021_1.iloc[0])
temp_2021_1.drop(temp_2021_1.index[0], inplace=True)
temp_2021_1
```

```python
temp_2021_2.drop(temp_2021_2.index[0], inplace=True)
temp_2021_2 = temp_2021_2.rename(columns=temp_2021_2.iloc[0])
temp_2021_2.drop(temp_2021_2.index[0], inplace=True)

temp_2021_3.drop(temp_2021_3.index[0], inplace=True)
temp_2021_3 = temp_2021_3.rename(columns=temp_2021_3.iloc[0])
temp_2021_3.drop(temp_2021_3.index[0], inplace=True)

temp_2022_1.drop(temp_2022_1.index[0], inplace=True)
temp_2022_1 = temp_2022_1.rename(columns=temp_2021_1.iloc[0])
temp_2022_1.drop(temp_2022_1.index[0], inplace=True)

temp_2022_2.drop(temp_2022_2.index[0], inplace=True)
temp_2022_2 = temp_2022_2.rename(columns=temp_2021_2.iloc[0])
temp_2022_2.drop(temp_2022_2.index[0], inplace=True)

temp_2022_3.drop(temp_2022_3.index[0], inplace=True)
temp_2022_3 = temp_2022_3.rename(columns=temp_2022_3.iloc[0])
temp_2022_3.drop(temp_2022_3.index[0], inplace=True)

temp_2023_1.drop(temp_2023_1.index[0], inplace=True)
temp_2023_1 = temp_2023_1.rename(columns=temp_2023_1.iloc[0])
temp_2023_1.drop(temp_2023_1.index[0], inplace=True)

temp_2023_2.drop(temp_2023_2.index[0], inplace=True)
temp_2023_2 = temp_2023_2.rename(columns=temp_2023_2.iloc[0])
temp_2023_2.drop(temp_2023_2.index[0], inplace=True)
```

```python
temp_2021_1
```

- 2022_1, 2022_2 컬럼명이 달라서 동일하게 맞추기

```python
temp_2022_1.columns = ['일자', '기온(°C)', '풍속(km/h)', '체감온도(°C)']
temp_2022_2.columns = ['일자', '기온(°C)', '풍속(km/h)', '체감온도(°C)']
```

### 2021, 2022, 2023년 데이터 만들기

```python
temp_2021 = pd.concat([temp_2021_1, temp_2021_2, temp_2021_3], axis = 0, ignore_index=True)
temp_2021.drop('습도(%rh)', axis=1, inplace=True)
temp_2021
```

```python
temp_2022 = pd.concat([temp_2022_1, temp_2022_2, temp_2022_3], axis = 0, ignore_index=True)
temp_2022
```

```python
temp_2023 = pd.concat([temp_2023_1, temp_2023_2], axis = 0, ignore_index=True)
temp_2023.drop('습도(%rh)', axis=1, inplace=True)
temp_2023
```

## 데이터 저장하기

```python
temp_2021.to_csv("체감온도_2021.csv", index=False)
temp_2022.to_csv("체감온도_2022.csv", index=False)
temp_2023.to_csv("체감온도_2023.csv", index=False)
```
