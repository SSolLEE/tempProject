(서울시 지하철역 정보 검색 (역명))

```python
import pandas as pd

df = pd.read_csv("[지하철_104]_서울시지하철역정보검색(역명).csv")
df.호선.value_counts()
```

```python
df = df[(df['호선'] == '01호선') | (df['호선'] == '02호선') | (df['호선'] == '03호선') |
        (df['호선'] == '04호선') | (df['호선'] == '05호선') | (df['호선'] == '06호선') |
        (df['호선'] == '07호선') | (df['호선'] == '08호선')]
df.호선.value_counts()
```
