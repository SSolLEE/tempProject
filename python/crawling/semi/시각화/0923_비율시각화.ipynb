# 역별 요일별 비율 시각화

```python
import pandas as pd
df = pd.read_csv("../data/0915_역별요일비율(결측치해결)_인천7빼고.csv")
data = pd.read_csv("../data/0915_역별월별비율_15_22.csv")
data.head(2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
import platform
print(platform.system())
# 운영체제별 한글 폰트 설정
if platform.system() == 'Darwin': # Mac 환경 폰트 설정
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows': # Windows 환경 폰트 설정
    plt.rc('font', family='Malgun Gothic')

plt.rc('axes', unicode_minus=False) # 마이너스 폰트 설정


# 글씨 선명하게 출력하는 설정
%config InlineBackend.figure_format = 'retina'
```

```python
data_cmr = data[data['지하철역'] == '성신여대입구']
pivot_df = data_cmr.pivot(index='사용월', columns='호선명', values='비율')

ax = pivot_df.plot(kind='bar', stacked=True, figsize=(10, 6), color='#3388FF')
plt.xlabel('월')
plt.ylabel('비율')
plt.title('4호선 성신여대입구역 월별 비율')
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
plt.legend(title='호선명')
plt.show()
```

```python
df_cmr = df[df['역명'] == '성신여대입구']
pivot = df_cmr.pivot(index='요일', columns='호선', values='비율')

ax = pivot.plot(kind='bar', stacked=True, figsize=(10, 6), color= '#3388FF')
plt.xlabel('요일')
plt.ylabel('비율')
plt.title('4호선 성신여대입구 요일별 비율')
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
plt.legend(title='호선명')
plt.show()
```
