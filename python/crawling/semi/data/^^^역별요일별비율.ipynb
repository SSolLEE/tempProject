## 이것저것 합 구하기
### 월요일 비율 = (월요일 이용객수) / {(일주일 전체 합)/7}

- 합계 컬럼 만들기

```python
data_on['합'] = data_on.iloc[:, 6:-2].sum(axis=1)
data_on.head(2)
```

```python
data_on[data_on.isna().any(axis=1)]
```

- 역별 요일 별 합계 데이터프레임 만들기 

```python
df = data_on.groupby(['호선', '역명', '요일'])['합'].mean().reset_index()

# 과학적 표기법 사용 해제
pd.options.display.float_format = '{:.2f}'.format
df[df['역명'] == '충무로']
### 여기서 합 컬럼은 각각의 요일 합의 평균값이다.
```

```python
# 충무로역은 3호선과 4호선의 통합역사이자 개찰구를 공동으로 사용하므로 모든 이용객은 4호선의 이용객으로 집계된다.
# 따라서 혼잡도 비율을 가지고 승객 수를 분리하고자 한다. 
##평일:  3호선 8.2 + 21.7 = 29.9  ## 4호선 12.0 + 37.3 = 49.3  ### 총 합 79.2
##토요일:  3호선 6.3 + 18.8 = 25.1  ## 4호선 10.5 + 26.3 = 36.8  ### 총 합 61.9
##일요일:  3호선 3.6 + 9.7 = 13.3  ## 4호선 6.1 + 13.7 = 19.8  ### 총 합 33.1
```

```python
value_4_0 = df[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '월')]['합'].values[0]
value_4_1 = df[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '화')]['합'].values[0]
value_4_2 = df[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '수')]['합'].values[0]
value_4_3 = df[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '목')]['합'].values[0]
value_4_4 = df[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '금')]['합'].values[0]
value_4_5 = df[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '토')]['합'].values[0]
value_4_6 = df[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '일')]['합'].values[0]
# 비율을 곱한 후 3호선 충무로 금요일 값에 할당
df.loc[(df['호선'] == 3) & (df['역명'] == '충무로') & (df['요일'] == '월'), '합'] = value_4_0 * (29.9/79.2)
df.loc[(df['호선'] == 3) & (df['역명'] == '충무로') & (df['요일'] == '화'), '합'] = value_4_1 * (29.9/79.2)
df.loc[(df['호선'] == 3) & (df['역명'] == '충무로') & (df['요일'] == '수'), '합'] = value_4_2 * (29.9/79.2)
df.loc[(df['호선'] == 3) & (df['역명'] == '충무로') & (df['요일'] == '목'), '합'] = value_4_3 * (29.9/79.2)
df.loc[(df['호선'] == 3) & (df['역명'] == '충무로') & (df['요일'] == '금'), '합'] = value_4_4 * (29.9/79.2)
df.loc[(df['호선'] == 3) & (df['역명'] == '충무로') & (df['요일'] == '토'), '합'] = value_4_5 * (25.1/61.9)
df.loc[(df['호선'] == 3) & (df['역명'] == '충무로') & (df['요일'] == '일'), '합'] = value_4_6 * (13.3/33.1)
```

```python
df.loc[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '월'), '합'] = df.loc[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '월'), '합'] * (49.3/79.2)
df.loc[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '화'), '합'] = df.loc[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '화'), '합'] * (49.3/79.2)
df.loc[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '수'), '합'] = df.loc[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '수'), '합'] * (49.3/79.2)
df.loc[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '목'), '합'] = df.loc[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '목'), '합'] * (49.3/79.2)
df.loc[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '금'), '합'] = df.loc[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '금'), '합'] * (49.3/79.2)
df.loc[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '토'), '합'] = df.loc[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '토'), '합'] * (36.8/61.9)
df.loc[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '일'), '합'] = df.loc[(df['호선'] == 4) & (df['역명'] == '충무로') & (df['요일'] == '일'), '합'] * (19.8/33.1)
```

```python
# 연신내역은 3호선이 있고, 6호선 순환선이 있다. 
# 따라서 혼잡도 비율을 가지고 승객 수를 분리하고자 한다. 
##평일:  3호선 25.2 + 23.4 = 48.6  ## 6호선 12.3   ### 총 합 60.9
##토요일:  3호선 19.9 + 10.3 = 30.2  ## 6호선 7.8  ### 총 합 38.0
##일요일:  3호선 9.6 + 5.5 = 15.1  ## 6호선 4.8  ### 총 합 19.9
```

```python
value_3_0 = df[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '월')]['합'].values[0]
value_3_1 = df[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '화')]['합'].values[0]
value_3_2 = df[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '수')]['합'].values[0]
value_3_3 = df[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '목')]['합'].values[0]
value_3_4 = df[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '금')]['합'].values[0]
value_3_5 = df[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '토')]['합'].values[0]
value_3_6 = df[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '일')]['합'].values[0]
# 비율을 곱한 후 6호선 연신내역에 할당
df.loc[(df['호선'] == 6) & (df['역명'] == '연신내') & (df['요일'] == '월'), '합'] = value_3_0 * (12.3/60.9)
df.loc[(df['호선'] == 6) & (df['역명'] == '연신내') & (df['요일'] == '화'), '합'] = value_3_1 * (12.3/60.9)
df.loc[(df['호선'] == 6) & (df['역명'] == '연신내') & (df['요일'] == '수'), '합'] = value_3_2 * (12.3/60.9)
df.loc[(df['호선'] == 6) & (df['역명'] == '연신내') & (df['요일'] == '목'), '합'] = value_3_3 * (12.3/60.9)
df.loc[(df['호선'] == 6) & (df['역명'] == '연신내') & (df['요일'] == '금'), '합'] = value_3_4 * (12.3/60.9)
df.loc[(df['호선'] == 6) & (df['역명'] == '연신내') & (df['요일'] == '토'), '합'] = value_3_5 * (7.8/38.0)
df.loc[(df['호선'] == 6) & (df['역명'] == '연신내') & (df['요일'] == '일'), '합'] = value_3_6 * (4.8/19.9)
```

```python
df.loc[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '월'), '합'] = df.loc[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '월'), '합'] * (48.6/60.9)
df.loc[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '화'), '합'] = df.loc[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '화'), '합'] * (48.6/60.9)
df.loc[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '수'), '합'] = df.loc[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '수'), '합'] * (48.6/60.9)
df.loc[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '목'), '합'] = df.loc[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '목'), '합'] * (48.6/60.9)
df.loc[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '금'), '합'] = df.loc[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '금'), '합'] * (48.6/60.9)
df.loc[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '토'), '합'] = df.loc[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '토'), '합'] * (30.2/38.0)
df.loc[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '일'), '합'] = df.loc[(df['호선'] == 3) & (df['역명'] == '연신내') & (df['요일'] == '일'), '합'] * (15.1/19.9)
```
