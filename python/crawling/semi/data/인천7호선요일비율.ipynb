```python
import pandas as pd

df1 = pd.read_csv("인천교통공사 1,2,7호선 역별일별 이용인원현황(20220630).csv", encoding="euc-kr")
df2 = pd.read_csv("인천교통공사_역별일별 이용인원현황_20230630.csv", encoding="euc-kr")
```

```python
df = pd.concat([df1, df2])
df
```

### 2022년 7호선 데이터만 추출

```python
df_7 = df[df['호선'] == 7]
```

```python
df_7.info()
```

```python
df7 = df_7[df_7['통행일자'].str.contains('2022')]
df7
```

### 요일 추가

```python
df7['통행일자'] = pd.to_datetime(df7['통행일자'])
df7['요일'] = df7['통행일자'].dt.weekday
df7.head(1)
```

```python
weekday_list = ['월', '화', '수', '목', '금', '토', '일']
df7['요일'] = df7.apply(lambda x: weekday_list[x['요일']], axis=1)
df7.head(1)
```

### 승차 컬럼만 추출

```python
df7 = df7[df7['구분'] == '승차']
df7
```

```python
df7.loc[df7['통행일자'] == '2022-01-01', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-01-31', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-02-01', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-02-02', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-03-01', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-03-09', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-05-05', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-05-08', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-06-01', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-06-06', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-08-15', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-09-09', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-09-10', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-09-11', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-09-12', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-10-03', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-10-09', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-10-10', '요일'] = '일'
df7.loc[df7['통행일자'] == '2022-12-25', '요일'] = '일'
df7
```

### 역별 요일별 합 평균

```python
df7_mean = df7.groupby(['호선', '역명', '요일'])['합계'].mean().reset_index()

# 과학적 표기법 사용 해제
pd.options.display.float_format = '{:.2f}'.format
df7_mean = df7_mean.replace({'역명' : '7호선 부평구청'}, '부평구청')
df7_mean
### 여기서 합 컬럼은 각각의 요일 합의 평균값이다.
```

```python
# 전체요일평균 / 7 
result = (df7_mean.groupby(['호선', '역명'])['합계'].sum() / 7 ).reset_index()
# result
df7_mean['비율'] = df7_mean['합계'] / df7_mean.apply(
    lambda x: result[(result['호선'] == x['호선']) & (result['역명'] == x['역명'])]['합계'].values[0], axis=1)
df7_mean.head(7)
```
