```python
import requests

url = "http://data.ex.co.kr/openapi/trtm/realUnitTrtm?key=1308370912&type=json&iStartUnitCode=101&iEndUnitCode=103"
req = requests.get(url)
print(req)
```

```python
json_df = req.json()
json_df
```

```python
## JSON 데이터를 Pandas 데이터프레임으로 가공하자.
```

```python
import pandas as pd

cars = json_df['realUnitTrtmVO']
pd.DataFrame(cars)
```

```python
records = []
for car in cars:
    row_dict = {}
    row_dict['data'] = car["stdDate"]
    row_dict['timeAvg'] = car['timeAvg']
    records.append(row_dict)
pd.DataFrame(records)
```

```python
import requests
url = "http://data.ex.co.kr/openapi/trtm/realUnitTrtm?key=1308370912&type=xml&iStartUnitCode=101&iEndUnitCode=103&numOfRows=10&pageNo=1"
req = requests.get(url)
print(req.text)
```

```python
## xml 결과값을 데이터프레임으로 변환
```

```python
import requests
import lxml
from bs4 import BeautifulSoup

url = "http://data.ex.co.kr/openapi/trtm/realUnitTrtm?key=1308370912&type=xml&iStartUnitCode=101&iEndUnitCode=103&numOfRows=10&pageNo=1"
req = requests.get(url)
# print(req.text)

soup = BeautifulSoup(req.text, "lxml") # XML 반환값이라서 lxml을 사용한 것
print(soup.prettify())

# xml결과값으로 데이터프레임 변환
stddate_list = soup.find_all("stddate")
timeAvg_list = soup.find_all("timeAvg")

stddate = []
avgTime = []
for date, timeAvg in zip(stddate_list, timeAvg_list):
    print(date.get_text(), timeAvg.get_text())
    stddate.append(date.get_text())
    avgTime.append(timeAvg.get_text())

df = pd.DataFrame({"data":stddate, "avgTime" : avgTime})
df
```
